```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

var= [1,2,3,4,5]
var1= [1,4,9,16,6]

z = pd.DataFrame({"var":var,"var1":var1})

sns.lineplot(x =var,y=var1,data=z)

plt.show()
```

```python

dataset = sns.load_dataset("penguins")

sns.lineplot(x ="island",y="species",data=dataset,hue="sex",style="sex",palette="Accent",
             markers=["o","*"],dashes=False,legend="full"   )
plt.show()
```
