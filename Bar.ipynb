```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

dataset = sns.load_dataset("penguins")
dataset
```

```python
sns.barplot(x="island",y="bill_length_mm",data=dataset,hue="sex")
plt.show()
```

```python
change_order=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=dataset,hue="sex",order=change_order
            ,hue_order=["Female","Male"],ci=1)
plt.show()
```

```python
sns.barplot(x="bill_length_mm",y="island",data=dataset,hue="sex",orient="h",ci=1)
plt.show()
```
